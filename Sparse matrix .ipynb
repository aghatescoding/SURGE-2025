```python
import numpy as np
import scipy.sparse as sparse
from scipy.stats import norm
import matplotlib.pyplot as plt
from IPython.display import Image, display

def sprandsym(n, density):
    rvs = norm().rvs
    X = sparse.random(n, n, density=density, data_rvs=rvs, random_state=42)
    upper_X = sparse.triu(X)
    result = upper_X + upper_X.T - sparse.diags(X.diagonal())
    return result

# Use smaller size and lower density for speed
M = sprandsym(50, 0.02)

plt.figure(figsize=(4, 4))
plt.spy(M, markersize=1)
plt.axis('off')
plt.tight_layout()
plt.savefig('Structurally_symmetric.jpg', format='jpg', dpi=100, bbox_inches='tight')
plt.close()

display(Image(filename='Structurally_symmetric.jpg'))
```

```python
import numpy as np
import scipy.sparse as sparse
from scipy.stats import norm
import matplotlib.pyplot as plt
from IPython.display import Image, display

def sprandsym(n, density):
    rvs = norm().rvs
    X = sparse.random(n, n, density=density, data_rvs=rvs, random_state=42)
    upper_X = sparse.triu(X)
    result = upper_X + upper_X.T - sparse.diags(X.diagonal())
    return result

# Generate the symmetric sparse matrix
M = sprandsym(100, 0.05)

plt.figure(figsize=(4, 4))
plt.spy(M, markersize=1)

# Set ticks at 0, 20, 40, 60, 80, 100
tick_positions = np.arange(0, 101, 20)
plt.xticks(tick_positions)
plt.yticks(tick_positions)

plt.tight_layout()
plt.savefig('Structurally_symmetric.jpg', format='jpg', dpi=100, bbox_inches='tight')
plt.close()

display(Image(filename='Structurally_symmetric.jpg'))
```
